```python
import pandas as pd
```

```python
# loading data
mh = pd.read_csv("../../data/raw/mentalhealthintech.csv")
mh
```

```python
list(mh.columns)
```

```python
mh_cleaned = mh.drop(['Timestamp', 'state', 'family_history', 'coworkers', 'Age', 'comments', 'no_employees', 'Country'], axis='columns')
mh_cleaned
```

```python
mh_cleaned = mh_cleaned[mh_cleaned['self_employed'].str.contains('Yes') == False]
mh_cleaned = mh_cleaned[mh_cleaned['tech_company'].str.contains('No') == False]
```

```python
mh_cleaned = mh_cleaned.replace(['Cis Male', 'Mail', 'Make', 'Male', 'Male ', 'Male-ish', 'Malr', 'Male (CIS)', 'Male', 'Man', 'something kinda male?', 'cis male', 'm', 'maile', 'male', 'msle'], 'M')
mh_cleaned = mh_cleaned.replace(['Femake', 'Female', 'Female ', 'Female (cis)', 'Woman', 'f', 'female'], 'F')
mh_cleaned = mh_cleaned.replace(['Enby', 'non-binary', 'Agender'], 'NB')
mh_cleaned = mh_cleaned.replace(['Female (trans)', 'Trans-female'], 'MTF')
mh_cleaned = mh_cleaned.replace(['Androgyne', 'Genderqueer', 'fluid', 'male leaning androgynous', 'queer/she/they'], 'F')
```

```python
mh_cleaned['Gender'].unique()
```

```python
mh_cleaned = mh_cleaned.dropna(axis=0, how='any')
mh_cleaned = mh_cleaned.dropna(subset=['work_interfere'], axis=0, how='any').reset_index()
```

```python
mh_cleaned = mh_cleaned.drop(['index', 'self_employed', 'tech_company'], axis=1)
```

```python
mh_cleaned.to_csv('mh_cleaned_gale.csv', index=None)
```
