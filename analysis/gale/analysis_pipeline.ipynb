```python
import pandas as pd
```

```python
# loading data
mh = pd.read_csv("../../data/raw/mentalhealthintech.csv")
mh
```

```python
list(mh.columns)
```

```python
mh_cleaned = mh.drop(['Gender', 'Timestamp', 'remote_work', 'phys_health_consequence', 'state', 'phys_health_interview', 'family_history', 'Age', 'comments', 'no_employees', 'Country'], axis='columns')
mh_cleaned
```

```python
mh_cleaned = mh_cleaned[mh_cleaned['self_employed'].str.find('Yes') == -1]
mh_cleaned = mh_cleaned[mh_cleaned['tech_company'].str.find('No') == -1]
```

```python
mh_cleaned = mh_cleaned.dropna(axis=0, how='any').reset_index()
```

```python
mh_cleaned = mh_cleaned.drop(['index', 'self_employed', 'tech_company'], axis=1)
```

```python
mh_cleaned.to_csv('mh_cleaned_gale.csv', index=None)
```
